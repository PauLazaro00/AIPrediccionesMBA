```python
import pandas as pd
import statsmodels.api as sm
from google.colab import files

# 1. Subir archivo manualmente desde tu PC
uploaded = files.upload()

# 2. Cargar el CSV en un DataFrame
df = pd.read_csv("datos_ventas_inf.csv")

# 3. Si existe columna fecha, convertirla
if "fecha" in df.columns:
    df["fecha"] = pd.to_datetime(df["fecha"], format="%Y-%m")

# 4. Definir variables independientes (X) y dependiente (y)
X = df[["inflacion", "emae"]]   # explicativas
y = df["venta"]                 # dependiente

# 5. Agregar constante (intercepto)
X = sm.add_constant(X)

# 6. Ajustar el modelo
modelo = sm.OLS(y, X).fit()

# 7. Mostrar resultados
print(modelo.summary())

from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# 1. Generar las predicciones del modelo
y_pred = modelo.predict(X)

# 2. Calcular MAE
mae = mean_absolute_error(y, y_pred)

# 3. Calcular RMSE
rmse = np.sqrt(mean_squared_error(y, y_pred))

# 4. Calcular MAPE (cuidado con ceros en ventas)
mape = (np.abs((y - y_pred) / y)).mean() * 100

# 5. Calcular BIAS
bias = (y_pred - y).mean()

print("MAE:", round(mae,2))
print("RMSE:", round(rmse,2))
print("MAPE:", round(mape,2), "%")
print("BIAS:", round(bias,2))

```
